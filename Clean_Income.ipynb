```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Ressources/Data/Camp_Market_Cleaned.csv", sep=",")
print(df.columns.tolist())
```

```python
print(pd.get_option("display.max_rows"))
print(pd.get_option("display.max_columns"))
```

```python
empty_rows = df[df['Income'].isna()]
display(empty_rows)
```

```python
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
```

```python
missing_idx = df['Income'].isna()  

def impute_income(row, df):
    if pd.notna(row['Income']):
        return row['Income']  # déjà rempli
    
    mask = (
        (df['Year_Birth'] == row['Year_Birth']) &
        (df['Education'] == row['Education']) &
        (df['Marital_Status'] == row['Marital_Status']) &
        (df['Income'].notna())
    )
    
    group_values = df.loc[mask, 'Income']
    
    if len(group_values) > 0:
        return round(group_values.mean())  # moyenne arrondie du groupe
    else:
        return np.nan  # pas de valeur connue dans le groupe

df['Income'] = df.apply(lambda row: impute_income(row, df), axis=1).astype('Int64')
df['Income_Imputed'] = missing_idx & df['Income'].notna()
```
